```python
import pandas as pd
```

```python
df = pd.read_csv("StudentHabitsandAcademicPerformanceDataset.csv")
df.head()
```

```python
df.isin(['', ' ', 'N/A', 'NA', 'null', 'None', 'Unknown']).sum()
```

```python
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].apply(lambda x: x.str.strip().str.lower())
df.head()
```

```python
df.info()
```

```python
binary_map_cols = [
    'part_time_job',
    'access_to_tutoring',
    'dropout_risk',
    'extracurricular_participation'
]
for col in binary_map_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0})
```

```python
ordinal_mappings = {
    'diet_quality': {'poor': 0, 'fair': 1, 'good': 2},
    'internet_quality': {'low': 0, 'medium': 1, 'high': 2},
    'family_income_range': {'low': 0, 'medium': 1, 'high': 2}
}
for col, mapping in ordinal_mappings.items():
    df[col] = df[col].map(mapping)
```

```python
df.head()
```

```python
float_cols = df.select_dtypes(include=['float']).columns

df[float_cols] = df[float_cols].round(1)
```

```python
nominal_columns = [
    'gender',
    'major',
    'parental_education_level',
    'study_environment',
    'learning_style'
]
df = pd.get_dummies(df, columns=nominal_columns, drop_first=True, dtype=int)
```

```python
df.to_csv("Student_Habits_Cleaned.csv", index=False)
```
